# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time
# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> apertar uma tecla
# pyantogui.hotkey-> apertar uma combinação de teclas(atalho)
# ( pyautogui, um pacote de codigos ja feito por outros programadores para automação de codigos)
# Interessante utilizar pausas nos comandos, pois depende da velocidade do site ou até mesmo do sistema para rodar e isso pode dar erro.

pyautogui.PAUSE = 1

# Passo 1: Acessar o Sistema da empresa
pyautogui.hotkey("ctrl","t")
pyautogui.write(" https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema")
pyautogui.press("enter")

time.sleep(5)

    
# Passo 2: Fazer Login no Sistema da empresa
pyautogui.click(x=955,y=463)
pyautogui.write("Rodrigo")
pyautogui.click(x=949,y=571)
pyautogui.write("123456")
pyautogui.click(x=985,y=651)
time.sleep(5)
# Passo 3: Baixar a Base de Dados Da Empresa (usar a tag clicks pode indicar a quantidade de clicks que voce vai fazer)
pyautogui.click(x=593, y=575, button="right")
pyautogui.click(x=793, y=899)

time.sleep(5)

    

    
```

```python
# Passo 4: Calcular os Indicadores, Que sao os pedidos Do chefe( sempre que for trabalhar com base de dados, o pandas é otimo pra isso, serve para qualquer tipo)
import pandas as pd

#Importar a base de dados
basededados = pd.read_csv(r"C:\Users\T-Gamer\Downloads\Compras.csv", sep=";") #Usar esse r minusculo antes pro phyton nao ficar perdido
display(basededados)
# calculo dos indicadores

# total gasto-> soma da coluna ValoFinal
total_gasto = basededados["ValorFinal"]. sum()

# quantidade-> soma da coluna quantidade
quantidade = basededados["Quantidade"]. sum()

# preço medio-> soma do valor final/ soma da quantidade
preço_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preço_medio)
```

```python
import pyperclip
#Passo 5: Enviar o E-mail para a diretoria/chefe

# entrar no meu email
pyautogui.hotkey("ctrl","t")
pyautogui.write("https://mail.yahoo.com/d/folders/1")
pyautogui.press("enter")
time.sleep(5)

#clicar em escrever
pyautogui.click(x=227, y=311)
pyautogui.write("guilhermomasid@gmail.com")
pyautogui.press("tab")
pyautogui.click(x=446, y=360)
pyperclip.copy("Relatório")
pyautogui.hotkey("ctrl","v")
pyautogui.press("tab")

#preencher o email ( o f antes das 3 aspas, serve para conseguir colocar variaveis dentro do texto modificavel)
texto= f"""
Prezados,

Segue o relatório de compras:

Total de Gastos: R${total_gasto:,.2f}

Quantidade de Produtos: {quantidade:,}

Preço Médio: R${preço_medio:,.2f}

Qualquer Dúvida é so falar
Att.

Rodrigo Abreu Amorim

"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")

#enviar
pyautogui.hotkey("ctrl","enter")
```
